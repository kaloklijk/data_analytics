# Update

## Dropping Rows

```python
# showing how many null values in columns
df.isnull().sum()
```

```python
# drop null value
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

# Update/drop/Add

```python
# update
s = df.copy()
s['Churn']  = True
# print all exception in s
s[s['Churn']==False]
# however, in the original df
df[df['Churn']==False].head()
```

```python
# add new column
s['minutes per call'] = s['Total day minutes'] / (s['Total day calls']+1)
```

```python
# update with slicing
s.iloc[1:5, s.columns.get_loc('Churn')] = False
s.head(6)
```

```python
# drop minutes per call column
s.drop('minutes per call', axis = 1, inplace = True)
s.head()
```

# Condition based Updating 

```python
# apply for conditional changes
df['Churn Binary'] = df['Churn'].apply({lambda x: 1 if x==True else 0})
```

# data visualization

```python
state  =  df.iloc[:, [df.columns.get_loc('State'), df.columns.get_loc('Churn')]].set_index('State', inplace = False)
state['Churn binary'] = state['Churn'].apply({lambda x: 1 if x==True else 0})
state.head()
```

```python
state.drop('Churn', axis = 1).head()
```

```python
# basic chart
state.head(20).plot()
```

```python
# bar chart plot
state.head(30).plot.bar()
```

```python
# horizontal bar chart plot
state.head(100).plot.barh()
```

```python
# data visualization with seaborn
plt.figure('Charges by State')
barplot(x='State', y='Total day minutes', data = df).set_title('Charges by State')
plt.show()
```

```python
# seaborn violin plot
plt.figure(figsize=(15, 6))
violinplot(x='Churn', y='Total day charge', data=df).set_title('Churn by Charges')
```

```python
# random linear regression
from random import *
new_df = pd.DataFrame({'number': np.arange(10), 'randint' : [randint(0, 100) for x in np.arange(10)]})
new_df
lmplot(x='number', y='randint', data=new_df)
```
